```python
train = train.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])
test = test.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])
```

```python
train.head()
```

```python
train.Sex = train.Sex == "female"
test.Sex = test.Sex == "female"
```

```python
train.head()
```

```python
train.describe()
```

```python
train.Age = train.Age.fillna(29.70).astype(np.float32)
test.Age = test.Age.fillna(29.70).astype(np.float32)
```

```python
train.describe()
```

```python
test.describe()
```

```python
test.Fare = test.Fare.fillna(0).astype(np.float32)
```

```python
test.describe()
```

```python
cat_vars = ['Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked']
cont_vars = ['Age', 'Fare']
dep = 'Survived'
```

```python
for variable in cat_vars: 
    train[variable] = train[variable].astype('category').cat.as_ordered()
    test[variable] = test[variable].astype('category').cat.as_ordered()
```

```python
train = train[cat_vars + cont_vars + [dep]].copy()
```

```python
test[dep] = 0
test = test[cat_vars + cont_vars + [dep]].copy()
```

```python
for variable in (dep, cont_vars): 
    train[variable] = train[variable].astype('float32')
    test[variable] = test[variable].astype('float32')
```

```python
apply_cats(test, train)
```

```python
df, y, nas, mapper = proc_df(train, dep, do_scale=True)
nas
```

```python
df_test, _, nas, mapper = proc_df(test, dep, do_scale=True, na_dict=nas, mapper=mapper)
nas
```

```python
len(df)
```

```python
val_idx = get_cv_idxs(len(df), val_pct=0.2)
len(val_idx)
```

```python
model = ColumnarModelData.from_data_frame(PATH, val_idx, df, y, cat_flds=cat_vars, bs=64, test_df=df_test)
```

```python
cat_sizes = [(cat, len(train[cat].cat.categories)) for cat in cat_vars]
cat_sizes
```

```python
embedding_sizes = [(size, min(50, (size+1)//2)) for _,size in cat_sizes]
embedding_sizes
```

```python
learner = model.get_learner(embedding_sizes, len(train.columns)-len(cat_vars),
                           0.04, 1, [100, 100], [0.1], y_range=(0,1))
```

```python
learner.crit = nn.BCEWithLogitsLoss()
```

```python
learner.lr_find()
```
